```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('tmdb_5000_movies.csv')
df.head ()
```

```python
df.info()
```

```python
plt.figure(figsize=(20,10))
plt.xticks(rotation='vertical')
sns.scatterplot(x='budget',y='genres',data=df)
```

```python
df.columns
```

```python
plt.figure (figsize=(5,5))
df.sort_values('revenue', axis=0, ascending=False, inplace=True)
plt.xticks(rotation='vertical')
sns.barplot(x=df['genres'][:10],y=df['revenue'][:10],data=df)
```

```python
plt.figure (figsize=(5,5))
df.sort_values('budget', axis=0, ascending=False, inplace=True)
plt.xticks(rotation='vertical')
sns.barplot(x=df['genres'][:10],y=df['budget'][:10],data=df)
```

```python
df.columns
```

```python
plt.figure (figsize=(5,5))
df.sort_values('vote_average', axis=0, ascending=False, inplace=True)
plt.xticks(rotation='vertical')
sns.barplot(x=df['genres'][:10],y=df['vote_average'][:10],data=df)
```

```python
df['vote_count'].value_counts()
```

```python
df.columns
```

```python
df['profit']=df['revenue']-df['budget']
df.head()
```

```python
plt.figure (figsize=(5,5))
df.sort_values('profit', axis=0, ascending=False, inplace=True)
plt.xticks(rotation='vertical')
sns.barplot(x=df['genres'][:10],y=df['profit'][:10],data=df)
```
